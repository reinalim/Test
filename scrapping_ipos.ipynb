```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
import re
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set the executable path and initialize the chrome browser in splinter (For Windows user)
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False) 
```

```python
# Visit the IPOs stocks site to grab the 2021 list.
url = 'https://stockanalysis.com/ipos/2020-list/'
browser.visit(url)
```

```python
# Set up the HTML parser & scrape the 2021 stat info box.
html = browser.html
infobox_2020 = soup(html, 'html.parser')
stat_2020info = infobox_2020.find('p', class_="ipocount").get_text()
stat_2020info
```

```python
# Set up the HTML parser.
html = browser.html
table_data1 = soup(html, 'html.parser')
```

```python
# Scrape the text data only from the <td> tags for the row data.
table_row = [x.get_text() for x in table_data1.find_all('td')]
table_row
```

```python
# Convert the scrapped table rows into Pandas DataFrame.
# DataFrame will put the list in one column which we will need to split later.
# Print head(30) to check the first 5 sets of row data from the site matches.

row_df = pd.DataFrame(table_row)
row_df.head(30)
```

```python
# Check the tail end of the DataFrame and drop the rows needed.
# Side content of the site also has some <td> tags for upcoming IPOs we dont want (not part of main table).
row_df.tail(30)
```

```python
# Drop row index 2880 onwards (this is the side content we don't want).
# Drop a row by index & check the tail end again to confirm the last data in the site.
IPO2020_cleaned = row_df.drop([2880,2881,2882,2883,2884,2885,2886,2887,2888,2889,2890,2891,2892,2893,2894])
IPO2020_cleaned.tail(6)
```

```python
# Export the DataFrame to a CSV file.
IPO2020_cleaned.to_csv(r'C:\Users\Lora B\Documents\Development\Final_Project\Stocks_Analysis_IPOs\IPO2020_cleaned.csv', index = False)
```

```python
# End the brower
browser.quit()
```
