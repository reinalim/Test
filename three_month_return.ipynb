```python
import pandas as pd
from pandas.tseries.offsets import DateOffset
```

```python
IPO_df = pd.read_csv('raw_dataset/IPO_Listing_2019_2020.csv')
IPO_df.head()
```

```python
IPO_df['IPO_DATE'] = pd.to_datetime(IPO_df['IPO_DATE'], format = '%Y-%m-%d')
```

```python
IPO_df['IPO_DATE']
```

```python
IPO_df['THREE_MTH_DATE'] = IPO_df['IPO_DATE'] + pd.DateOffset(months=3)
```

```python
# Create a list of IPO symbols from the IPO_df
IPO_list = IPO_df['SYMBOL'].tolist()
IPO_list
```

```python
IPO_df['THREE_MTH_RETURN'] = '0'
```

```python
for symbol in IPO_list:
    complete_file_name = f'raw_dataset/daily_adjusted/{symbol}_daily_adjusted.csv'
    ticker_df = pd.read_csv(complete_file_name)
    try:
        ticker_df['timestamp'] = pd.to_datetime(ticker_df['timestamp'], format = '%Y-%m-%d')
        ticker_df = ticker_df.set_index('timestamp')
        ticker_date = IPO_df['THREE_MTH_DATE'][symbol]
        index_ticker = ticker_df.index.get_loc(ticker_date.strftime('%Y-%m-%d'))
        three_mth_return_value = ticker_df['close'][index_ticker.item()]
        IPO_df.THREE_MTH_RETURN.loc[symbol] = three_mth_return_value
    except:
        pass
    
```

```python
IPO_df.rename(columns={'NAME':'COMPANY_NAME', 'CURRENT':'CURRENT_PRICE','RETURN':'PERCENTAGE_RETURN'}, inplace=True)
```

```python
# Export to csv file
IPO_df.to_csv('raw_dataset/three_month_return.csv')
```
